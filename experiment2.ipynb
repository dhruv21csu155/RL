```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
x = np.random.randint(0,50,1000)
y = x + np.random.randint(0,10,1000)
a = np.corrcoef(x,y)
print(a)
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
x = np.random.randint(0,50,1000)
y = 100-x + np.random.randint(0,10,1000)
a = np.corrcoef(x,y)
print(a)
plt.scatter(x,y)
plt.show()
```

```python
import pandas as pd
df= pd.DataFrame({'a' : np.random.randint(0,50,1000)})
df['b']= df['a'] + np.random.normal (0,10,1000)
df['c']= 100-df['a'] + np.random.normal(0,5,1000)
df['d']= np.random.randint(0,50,1000)

df.corr()
pd.plotting.scatter_matrix(df, figsize=(6,6))
plt.show
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
